```python
import pandas as pd
import numpy as np
```

```python
forum_users = {
    'UserID': np.array([1, 2, 3, 4, 5]),
    'Username': ['jane_smith', 'alex123', 'bob56', 'mark_wilson', 'test_name'],
    'Age': [18, 35, 25, 28, None],
    'Joined Date': pd.to_datetime(['2034-01-01', '2034-02-15', '2034-04-25', '2034-06-21', '2034-09-15']),
    'Total Posts': [150, 230, 80, 420, 310],
    'Reputation': [500, 720, 200, 940, 500]
}

df = pd.DataFrame(forum_users)
df
```

```python
df['Language'] = 'English'
df
```

```python
df['Reputation'] = 100
df
```

```python
df['Reputation'] = [100, 200, 300, 400, 500]
df
```

```python
df['Active'] = np.array([True, False, True, True, False])
df
```

```python
total_comments = {
    'Total Comments': [70, 30, 45, 55, 80]
}
df2 = pd.DataFrame(total_comments)
df2
```

```python
# By default  concat works on the Axis 0!
df = pd.concat([df, df2], axis=1)
df
```

```python
# Original df doesn't change
df.drop(['Language'], axis=1)
df
```

```python
# Original df is change in place
df.drop(['Language'], axis=1, inplace=True)
df
```
